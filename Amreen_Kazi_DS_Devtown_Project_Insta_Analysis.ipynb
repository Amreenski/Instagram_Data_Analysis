```python
insta_zero=insta[insta['Impressions'] == 0]
```

```python
#we observe that all the null cells belong to the same rows
#all these rows have impressions 0
#also when impressions is 0 insta_views, fb_views and profile_visits is also 0
#hence filling NaN values with 0
```

```python
insta=insta.fillna(0)
```

```python
insta.isnull().sum()
```

```python
#converting date to datetime datatype
```

```python
insta.info()
```

```python
insta.date=pd.to_datetime(insta['date'],errors='coerce')
```

```python
insta
```

```python
insta.info()
```

```python
#extracting hours from time column
```

```python
lst=insta['time']
```

```python
lst
```

```python
lst2=[]
for i in lst:
   lst2.append(i[:2])
lst2
```

```python
insta['hour_time']=lst2
```

```python
insta
```

```python
insta.drop('time',axis=1,inplace=True)
```

```python
insta
```

```python
#OneHotEncoding
```

```python
lst=[]
for i in insta['multiple']:
    if i=='Yes':
        lst.append(1)
    else:
        lst.append(0)
```

```python
lst
```

```python
insta['multiple_posts']=lst
```

```python
insta.drop('multiple',axis=1,inplace=True)
```

```python
insta
```

```python
#extracting weekday from date column
```

```python
lst=[]
for i in insta['date']:
    a=i.weekday()
    if a==0:
        lst.append('Monday')
    elif a==1:
        lst.append('Tuesday')
    elif a==2:
        lst.append('Wednesday')
    elif a==3:
        lst.append('Thursday')
    elif a==4:
        lst.append('Friday')
    elif a==5:
        lst.append('Saturday')
    else:
        lst.append('Sunday')
```

```python
lst
```

```python
insta['day']=lst
```

```python
insta
```

```python
insta.drop('date',axis=1,inplace=True)
```

```python
insta
```

```python
#visualisation and further expl0ratory analysis for better understanding
```

```python
def visualization(column1,column2):
    plt.figure(figsize=(12,5))
    plt.scatter(insta[column1],insta[column2])
    plt.ylabel(column2)
    plt.xlabel(column1)
    plt.grid()
```

```python
visualization('hour_time','reach')
```

```python
def visualization2(column1,column2):
    plt.figure(figsize=(18,5))
    sns.lineplot(insta[column1],insta[column2])
    plt.grid()
```

```python
visualization2('hour_time','reach')
```

```python
visualization('day','reach')
```

```python
visualization('type','reach')
```
